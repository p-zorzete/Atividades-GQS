```
%%writefile is_par.py
def is_even(value: int) -> bool:
    """Retorna True se o valor for par."""
    return value % 2 == 0
```

```
%%writefile test_is_par.py
import unittest
from is_par import is_even

class TestIsEven(unittest.TestCase):
    def test_even_number(self):
        self.assertTrue(is_even(10))

    def test_odd_number(self):
        self.assertFalse(is_even(7))

    def test_zero(self):
        self.assertTrue(is_even(0))

    def test_negative_even(self):
        self.assertTrue(is_even(-14))

    def test_negative_odd(self):
        self.assertFalse(is_even(-3))

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```
!pytest -q test_is_par.py
```

```
%%writefile fatorial.py
def fatorial(n: int) -> int:
    """Calcula o fatorial de n."""
    if not isinstance(n, int):
        raise TypeError("n deve ser um inteiro")
    if n < 0:
        raise ValueError("n deve ser >= 0")

    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado
```

```
%%writefile test_fatorial.py
import unittest
from fatorial import fatorial

class TestFatorial(unittest.TestCase):

    def test_zero(self):
        self.assertEqual(fatorial(0), 1)

    def test_pequeno_positivo(self):
        self.assertEqual(fatorial(4), 24)

    def test_valor_maior(self):
        self.assertEqual(fatorial(6), 720)

    def test_negativo_invalido(self):
        with self.assertRaises(ValueError):
            fatorial(-5)

    def test_tipo_invalido(self):
        with self.assertRaises(TypeError):
            fatorial(3.5)

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```
!pytest -q test_fatorial.py
```

```
%%writefile conta.py
class SaldoInsuficiente(Exception):
    """Erro lançado quando o saldo não cobre o saque."""
    pass

class Conta:
    def __init__(self, saldo=0):
        if saldo < 0:
            raise ValueError("Saldo inicial não pode ser negativo")
        self.saldo = saldo

    def depositar(self, valor):
        if valor <= 0:
            raise ValueError("Depósito deve ser positivo")
        self.saldo += valor

    def sacar(self, valor):
        if valor <= 0:
            raise ValueError("Saque deve ser positivo")
        if valor > self.saldo:
            raise SaldoInsuficiente("Saldo insuficiente")
        self.saldo -= valor
```

```
%%writefile test_conta.py
import unittest
from conta import Conta, SaldoInsuficiente

class TestConta(unittest.TestCase):

    def setUp(self):
        self.c = Conta(200)

    def test_deposito_valido(self):
        self.c.depositar(100)
        self.assertEqual(self.c.saldo, 300)

    def test_saque_valido(self):
        self.c.sacar(50)
        self.assertEqual(self.c.saldo, 150)

    def test_saldo_insuficiente(self):
        with self.assertRaises(SaldoInsuficiente):
            self.c.sacar(500)

    def test_deposito_invalido(self):
        with self.assertRaises(ValueError):
            self.c.depositar(0)

    def test_saque_invalido(self):
        with self.assertRaises(ValueError):
            self.c.sacar(-20)

    def test_saldo_inicial_invalido(self):
        with self.assertRaises(ValueError):
            Conta(-10)

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```
!pytest -q test_conta.py
```

```
%%writefile clima.py
import requests

def buscar_clima(cidade: str) -> float:
    """Retorna a temperatura da cidade pela API."""
    url = f"https://api.exemplo/clima?cidade={cidade}"
    r = requests.get(url)
    data = r.json()

    if not isinstance(data, dict):
        raise ValueError("Resposta inesperada da API")

    temperatura = data.get("temperatura")
    if temperatura is None:
        raise ValueError("Resposta inválida da API: sem temperatura")

    return temperatura
```

```
%%writefile test_clima.py
from unittest.mock import patch, Mock
from clima import buscar_clima
import pytest

@patch("clima.requests.get")
def test_clima_ok(mock_get):
    mock = Mock()
    mock.json.return_value = {"temperatura": 22.5}
    mock_get.return_value = mock

    assert buscar_clima("Curitiba") == 22.5

@patch("clima.requests.get")
def test_clima_sem_campo(mock_get):
    mock = Mock()
    mock.json.return_value = {}
    mock_get.return_value = mock

    with pytest.raises(ValueError):
        buscar_clima("Rio")

@patch("clima.requests.get")
def test_resposta_malformada(mock_get):
    mock = Mock()
    mock.json.return_value = None
    mock_get.return_value = mock

    with pytest.raises(ValueError):
        buscar_clima("São Paulo")
```

```
!pytest -q test_clima.py
```
